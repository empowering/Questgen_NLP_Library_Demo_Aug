```
# 파일 불러오기
import pandas as pd
df = pd.read_csv("/content/drive/My Drive/SDP/final_result.tsv", delimiter = '\t')
df = df['body']
```

# Generate MCQ Questions


```
import csv
from pprint import pprint
from Questgen import main

qg = main.QGen()
```

```
import json
file_path = "QA_output.json"
data = {}
```

```
number = 0
with open(file_path, 'w') as outfile:
    for each in df:
        number += 1
        print("="*10, number, "="*10)

        inputs = {} 
        inputs["input_text"] = each
        try:
            output = qg.predict_shortq(inputs)
            data[number] = []
            data[number].append(output)
        except TypeError :
            data[number] = []
        json.dump(data, outfile, indent=4)
        
```
